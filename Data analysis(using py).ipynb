```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #visualizaion data
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
#encoding error to avoid we use 'unicode_escape'
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df.info()
```

```python
df['Amount'].dtype
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

# Gender

```python
ax=sns.countplot(x="Gender",data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)


sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

## The number of orders of female is much comparitive to Male

```python
print(sales_gen)
```

# Age

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
 sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)


sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

### The above graph represents the hightest number of order are getting from 26-35 that too the gender Female

# State

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sns.set(rc={'figure.figsize':(20,5)})
sales_State=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)



sns.barplot(x='State',y='Amount',data=sales_State)
```

### For above graph we can see that most of the orders are from uttarprdesh,Maharastra and Karnataka respectively but the amount values difference if we see the state like Haryana and Kerala 

# Marital_status

```python
df.columns
```

```python
ax=sns.countplot(data=df,x='Marital_Status')

sns.set(rc={'figure.figsize':(8,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Marital_status=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(8,5)})

sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_Marital_status)
```

### From the above graph we can see that most of the buyers are married (women) and they have high purchasing order.

# Occupation

```python
sns.set(rc={'figure.figsize':(20,8)})
ax=sns.countplot(data=df,x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python

sns.set(rc={'figure.figsize':(20,5)})
sales_Occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)



sns.barplot(x='Occupation',y='Amount',data=sales_Occupation)
```

### From the above graph the most of the buyers from IT sector followed by healthcare,Aviation sectors.

# Product_category

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(22,8)})
ax=sns.countplot(data=df,x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```
